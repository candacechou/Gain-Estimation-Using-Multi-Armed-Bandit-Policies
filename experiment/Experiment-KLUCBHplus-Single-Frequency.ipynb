# KLUCBplus with Single Frequency Strategy

$\sigma$ = 0.1


```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
from bandit import Bandits
from KLUCB import KLUCB_Single_Freq
from Result_Visualization import System_Gain_Estimation,Plot_mean_square_error
from Regret import Model_Regret
from Model_Analysis import  Plot_Chosen_Time,System_Gain_Estimation_UCB
```

## Experiment 1

```python
B     = [0.1023, -0.1050, -0.0734, -0.2553, -0.0331, 0.2257, -0.0607, -0.2380, -0.0664, -0.1164, 0.1056]
A     = [1]
K     = 50
N     = 100000
sigma = 0.1
name  = "KLUCB-H-Plus"
```

### Create a bandit Object

```python
bandit = Bandits(A,B,K)

bandit.print_hinf_norm()
bandit.plot_real_system()
c = 3
Model_UCBSF = KLUCB_Single_Freq(bandit,sigma,N,name,c,"1")
```

### Before starting the experiment, the prior of the gain looks like ...

```python
System_Gain_Estimation_UCB(Model_UCBSF)
Plot_Chosen_Time(Model_UCBSF)
```

### Run the experiments

We run the experiments 50 times to calculate the emperical rewards and the regrets.

```python
Kn = 50
savefig = True
frame = 3000
Regret_UCBSF = Model_Regret(Model_UCBSF,Kn)
Regret_UCBSF.Compute_Regret(savefig=savefig,frame=frame)

```

### After experiment, the estimation of the gain looks like ...

```python
System_Gain_Estimation_UCB(Model_UCBSF)
Plot_Chosen_Time(Model_UCBSF)
```

### Plot the regret over 50 experiments.

```python
plt.figure()
for i in range(Kn): 
    plt.plot(range(len(Regret_UCBSF.regret[:,i])),Regret_UCBSF.regret[:,i])
plt.xlabel("Time")
plt.ylabel("Regret")
plt.title("The regret of each model.")
plt.show()
  
        
        
```

### Plot the Expected regret

```python
Regret_UCBSF.Plot_Regret()
```

### Plot the Mean Square Error

```python
Regret_UCBSF.Plot_MSE()
```

### Plot Regret with Regret lower bound on in semi-log

```python
Regret_UCBSF.Plot_Regret(T = True)
```

### Plot semilog MSE

```python
Regret_UCBSF.Plot_MSE()
```

### Save the Experiment 1 Result for Further Analysis

```python
import pickle
with open('./Result/KLUCBHplusSF_50arms_1e5rounds_EX1.pkl', 'wb') as handle:
    pickle.dump(Regret_UCBSF, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
