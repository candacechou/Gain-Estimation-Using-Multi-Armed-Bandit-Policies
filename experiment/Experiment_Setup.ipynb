# Experiment Setup

This notebook shows two of the experiment setup in the thesis. 

```python
from scipy import signal
from scipy.optimize import minimize
import numpy as np
import matplotlib.pyplot as plt
from bandit import Bandits
```

## Experiment 1

### Plot the frequency response of the system G.

```python
##### experiment 1 
B     = [0.1023, -0.1050, -0.0734, -0.2553, -0.0331, 0.2257, -0.0607, -0.2380, -0.0664, -0.1164, 0.1056]
A     = [1]
K     = 50
bandit = Bandits(A,B,K)
bandit.print_hinf_norm()
bandit.plot_real_system()
```

### Discretize the frequencies into 50 arms. (K = 50)

```python
plt.figure(figsize=(20/2.54,10/2.54))
plt.plot(bandit.arms,bandit.norm,color = 'black',label = '|$G_1$($e^{j\omega}$)|')
plt.scatter(bandit.arms,bandit.norm,color = 'red',label = '|$G_1$($e^{j\omega_k}$)|')
plt.title("Discretized Frequency Response of the system $G_1$ ")
plt.xlim([-0.1,3.2])
plt.ylim([-0.1,1])
plt.legend()
plt.xlabel('$\omega$ [rad/s]')
plt.ylabel("Amplitude")
plt.show()
```
