# KLUCBplus with Power Allocation Strategy

$\sigma$ = 0.1


```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
from bandit import Bandits
from KLUCB import KLUCB_Power_Allocation
from Result_Visualization import System_Gain_Estimation,Plot_mean_square_error
from Regret import Model_Regret
from Model_Analysis import Plot_Chosen_Time,System_Gain_Estimation_UCB
```

## Experiment 1

```python
B     = [0.1023, -0.1050, -0.0734, -0.2553, -0.0331, 0.2257, -0.0607, -0.2380, -0.0664, -0.1164, 0.1056]
A     = [1]
K     = 50
N     = 100000
sigma = 0.1
beta  = 1
c = 3
name  = 'KLUCB-Plus'
```

### Giving Uniform distribution of each arm

```python
prior = [np.random.random() for i in range(K)]
prior /= np.sum(prior)
```

### Create a bandit Object

```python
bandit = Bandits(A,B,K)

bandit.print_hinf_norm()
bandit.plot_real_system()

Model_KLUCBPA = KLUCB_Power_Allocation(bandit,sigma,N,name,prior,c,"1")
```

### Before starting the experiment, the prior of the gain looks like ...

```python
System_Gain_Estimation_UCB(Model_KLUCBPA)
Plot_Chosen_Time(Model_KLUCBPA)
```

### Run the experiments

We run the experiments 50 times to calculate the emperical rewards and the regrets.

```python
Kn = 50
savefig = True
frame = 5000
Regret_KLUCBPA = Model_Regret(Model_KLUCBPA,Kn)
Regret_KLUCBPA.Compute_Regret(showed=False,savefig = savefig,frame=frame)

```

### Plot the regret over 50 experiments.

```python
plt.figure()
for i in range(Kn): 
    plt.plot(range(len(Regret_KLUCBPA.regret[:,i])),Regret_KLUCBPA.regret[:,i])
plt.xlabel("Time")
plt.ylabel("Regret")
plt.title("The regret of each model.")
plt.show()
  
        
        
```

### Plot the Expected regret

```python
Regret_KLUCBPA.Plot_Regret()
```

### Plot the Mean Square Error

```python
Regret_KLUCBPA.Plot_MSE()
```

### Plot Regret with Regret lower bound on in semi-log

```python
Regret_KLUCBPA.Plot_Regret(T = True)
```

### Plot semilog MSE

```python
plt.figure()
plt.semilogy(Regret_KLUCBPA.MSE,label = 'MSE1')
plt.plot(Regret_KLUCBPA.MSE1,label = 'MSE2')
plt.title("Mean Square Error of the experiment.")
plt.xlabel("Time t")
plt.ylabel("MSE")

plt.legend()
plt.show()
```

### After experiment, the estimation of the gain looks like ...

```python
System_Gain_Estimation_UCB(Model_KLUCBPA)
Plot_Chosen_Time(Model_KLUCBPA)
```

### Save the Experiment 1 Result for Further Analysis

```python
import pickle
with open('./Result/KLUCBplusPA_50arms_1e5rounds_EX1.pkl', 'wb') as handle:
    pickle.dump(Regret_KLUCBPA, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
